```python
# Libraries

import pandas as pd
import numpy as np
```

```python
# Load data

data = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DL0101EN/labs/data/concrete_data.csv')
data.head()
```

```python
# Size of the dataset

data.shape
```

```python
# Summarize dataset

data.describe()
```

```python
# Are there missing values?

data.isnull().sum()
```

```python
# "Strength" is our target column
# All other columns are our predictors columns

data_columns = data.columns

predictors = data[data_columns[data_columns != 'Strength']]

target = data['Strength']
print(target.head())

# Normalizing Predictors and printing the first five rows.
predictors_norm = (predictors - predictors.mean()) / predictors.std()
print(predictors_norm.head())
```

```python
# number of predictors

n_cols = predictors_norm.shape[1]
n_cols
```

**Start building the model**

```python
import keras

#Other functions
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Define

def regression_model():
    #create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
#Split dataset to train_set and test_set
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=1)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
# Build the model

model_C = regression_model()
```

```python
# Train a model on the training data using 100 epochs
# Repeat Part B but use 100 epochs this time for training.

model_C.fit(X_train, y_train, epochs=50, verbose=2)
```

```python
#Evaluate the model
loss_val = model_C.evaluate(X_test, y_test)
y_pred = model_C.predict(X_test)
loss_val
```

```python
# Evaluate the model on the test data
from sklearn.metrics import mean_squared_error

mean_square_error = mean_squared_error(y_test, y_pred)
mean = np.mean(mean_square_error)
standard_deviation = np.std(mean_square_error)
print(mean, standard_deviation)
```
